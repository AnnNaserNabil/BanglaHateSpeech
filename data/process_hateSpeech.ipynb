```python
import pandas as pd

# 1) Row-wise merge
df_all = pd.concat([df, df2, dfx, df4, df5, df6, df7], ignore_index=True)

# 2) Make sure the key columns exist and align
expected_cols = ['Comments', 'HateSpeech']
for col in expected_cols:
    if col not in df_all.columns:
        df_all[col] = pd.NA

# 3) Coerce HateSpeech to 0/1 ints (keep NA if unexpected values)
df_all['HateSpeech'] = (
    pd.to_numeric(df_all['HateSpeech'], errors='coerce')
      .astype('Int64')
      .fillna(0)
      .astype(int)
)

# 4) Sort so 1s are first, 0s at the bottom (stable keeps original order within each group)
df_all = df_all.sort_values('HateSpeech', ascending=False, kind='stable').reset_index(drop=True)

# (optional) Reorder columns, drop exact duplicates, and quick check
df_all = df_all[[c for c in expected_cols if c in df_all.columns] + [c for c in df_all.columns if c not in expected_cols]]
df_all = df_all.drop_duplicates(subset=['Comments', 'HateSpeech'])
print(df_all['HateSpeech'].value_counts())
```

```python
df_all.to_csv('merged_hatespeech.csv', index=False, encoding='utf-8')
```
